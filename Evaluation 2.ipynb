```python
from rouge_score import rouge_scorer
```

```python
import json
with open('generated_summaries.json', 'r') as f:
    generated_summaries = json.load(f) # a list of LLM generated summaries
```

```python
with open('ground_truth.json', 'r') as f:
    ground_truth = json.load(f)[:1000] # a list of summaries provided by the original dataset
```

```python
scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)
```

```python
results = []
for i in range(1000):
    scores = scorer.score(ground_truth[i], generated_summaries[i])
    result = {}
    for k, v in scores.items():
        result[k+' P'] = v[0]
        result[k+' R'] = v[1]
        result[k+' F'] = v[2]
    results.append(result)
```
